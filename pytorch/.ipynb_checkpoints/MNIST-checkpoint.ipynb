```python
class MLPMnist(nn.Module):
    def __init__(self):
        super(MLPMnist,self).__init__()
        self.fc1 = nn.Linear(28*28, 50)
        self.fc2 = nn.Linear(50, 50)
        self.fc3 = nn.Linear(50, 10)
        
    def forward(self,x):
        x = x.view(-1, 28 * 28)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        return F.softmax(x,axis=1)
        
```

```python
model = MLPMnist().to(device)
```
